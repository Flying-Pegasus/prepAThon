## <h2><b>2.1</b> Feature Engineering</h2>

<h4><b>2.1a</b> Combine the approach_date, month, and year features into a single feature representing

```python
#Ratio of Miss Distance vs. Semi-major axis
data['MissDist/Semimajoraxis'] = data['Miss Dist.(kilometers)']/data['Semi Major Axis']
data['MissDist/Semimajoraxis'].head(4)
```

```python
# Create a 'Time Until Approach' feature based on the difference between the 'Epoch Date Close Approach' and the current date.

#Epoch date (converted) - Current Epoch time (-ve sign indicates that the day has passed)
data['Epoch Date Close Approach'] = pd.to_datetime(data['Epoch Date Close Approach'], unit='ms')
current_date = pd.to_datetime(datetime.now())
data['Time Until Approach (days)'] = (data['Epoch Date Close Approach'] - current_date).dt.days #Unit = days
data[['Epoch Date Close Approach', 'Time Until Approach (days)']].head(4)
```

<h4><b>2.1c</b> Calculate the eccentricity of the orbit, average orbital velocity, and orbital period using
Keplerâ€™s Law.</h4>

```python
# Constants
G = 6.67430e-11  # Gravitational constant 
M_sun = 1.989e30  # Mass of the Sun in kg
AU_to_meters = 1.496e11  # Conversion factor from AU to meters

# Ensure 'Semi Major Axis' and 'Aphelion Dist' are in AU, convert to meters

# 1. Calculate Eccentricity (e)
# e = (Aphelion Dist / Semi Major Axis)-1
eccentricity_threshold_lo = 0.000011 #threshold for circular orbits
eccentricity_threshold_hi = 0.999 # Slightly below 1
data['Eccentricity'] = ((data['Aphelion Dist'] * AU_to_meters) - (data['Semi Major Axis'] * AU_to_meters)) / (data['Semi Major Axis'] * AU_to_meters)
data['Eccentricity'] = data['Eccentricity'].clip(lower=eccentricity_threshold_lo, upper=eccentricity_threshold_hi)


# 2. Calculate Average Orbital Velocity (v_avg) in km/s
data['Average Orbital Velocity (km/s)'] = (np.sqrt(G * M_sun / (data['Semi Major Axis'] * AU_to_meters)))/1000

# Convert Orbital Period to years
data['Orbital Period (years)'] = (2 * np.pi * np.sqrt((data['Semi Major Axis'] * AU_to_meters)**3 / (G * M_sun))) / (60 * 60 * 24 * 365.25)

data[['Eccentricity','Average Orbital Velocity (km/s)','Orbital Period (years)']].head(4)
```

<h4><b>2.1d</b> Calculate the heliocentric distance, escape velocity, and specific orbital energy.</h4>

```python
# Calculate Heliocentric Distance
data['Heliocentric Distance (meters)'] = (data['Semi Major Axis'] * AU_to_meters)/1000

# Calculate Escape Velocity (v_e)
data['Escape Velocity (km/s)'] = (np.sqrt((2 * G * M_sun) / data['Heliocentric Distance (meters)']))/10e3

# Calculate Specific Orbital Energy (Îµ)
data['Specific Orbital Energy (MJ/kg)'] = - ((G * M_sun) / (2 * (data['Semi Major Axis'] * AU_to_meters))/10e6)

data[['Heliocentric Distance (meters)','Escape Velocity (km/s)','Specific Orbital Energy (MJ/kg)']].head(4)
```

<h4><b>2.1e</b> Calculate the Specific Angular Momentum using the formula: h=sqrt(GMa(1âˆ’e2)).</h4>

```python
# Calculate Specific Angular Momentum (h)
data['Specific Angular Momentum (m^2/s)'] = np.sqrt(G * M_sun * (data['Semi Major Axis'] * AU_to_meters) * (1 - data['Eccentricity'] ** 2))
data['Specific Angular Momentum (m^2/s)'].head(4)
```

<h4><b>2.1f</b> Calculate the velocity at Perihelion and Aphelion.</h4>

```python
# Calculate distance at Perihelion (r_p) and Aphelion (r_a)
data['Perihelion Distance (m)'] = (data['Semi Major Axis'] * AU_to_meters) * (1 - data['Eccentricity'])

# Calculate velocity at Perihelion (v_p)
data['Velocity at Perihelion (km/s)'] = (np.sqrt(G * M_sun* (2 / data['Perihelion Distance (m)'] - 1 / (data['Semi Major Axis'] * AU_to_meters))))/1000

# Calculate velocity at Aphelion (v_a)
data['Velocity at Aphelion (km/s)'] = (np.sqrt(G * M_sun * (2 / (data['Aphelion Dist'] * AU_to_meters) - 1 / (data['Semi Major Axis'] * AU_to_meters))))/1000

data[['Perihelion Distance (m)','Velocity at Perihelion (km/s)','Velocity at Aphelion (km/s)']].head(4)
```

<h4><b>2.1h</b> Calculate Synodic Period and Mean Motion using the orbital period.</h4>

```python
# 'Orbital Period' in years

# Constants
T_Earth = 1  # Orbital period of Earth in years

# Calculate Mean Motion (n) in degrees per day using orbital period year
data['Mean Motion (degrees/day)'] = 360 / (data['Orbital Period (years)'] * 365.25)

# Calculate Synodic Period (S) in years
# Using the formula S = 1 / |(1/T) - (1/T_Earth)|
data['Synodic Period (years)'] = 1 / (abs((1 / data['Orbital Period']) - (1 / T_Earth)))

data['Synodic Period (years)'].head(4)
```
